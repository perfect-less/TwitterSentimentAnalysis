```python
import pandas as pd
    
```

```python
#configuration
config = twint.Config()
config.Search = "#AbrahamAccords" #"#AbrahamAccords"
config.Lang = "en"
config.Limit = 1000
config.Min_likes = 50
# config.Year = "2021"
#config.Store_object = True
config.Pandas = True
# config.Since = "2020-09-14"
config.Until = "2020-09-16"
# config.Store_json = True
# config.Output = output_file
config.Hide_output = True
```

```python
current_date = "2020-09-15"
loop_count = 50

tweet_df = pd.DataFrame()
initiated = False

for i in range(loop_count):
    twint.run.Search(config)

    print (twint.output.panda.Tweets_df.shape[0])
    if twint.output.panda.Tweets_df.shape[0] > 0:
        if initiated == False:
            tweet_df = twint.output.panda.Tweets_df
            initiated = True
            continue
        else:
            tweet_df = pd.concat([tweet_df, twint.output.panda.Tweets_df], ignore_index=True)

tweet_df = tweet_df.drop_duplicates(subset=["id", "tweet"], ignore_index=True)
tweet_df = tweet_df.loc[tweet_df["date"] > current_date].reset_index(drop=True)
```
